```python
import os
import random
from PIL import Image
```

```python
path = "/Users/jinsungpark/Desktop/crypto_total"
file_lst = os.listdir(path)
file_lst.sort(key=len)
file_lst.remove('.DS_Store')
```

```python
lst_total = []
lst_0to9 = []
lst_10to99 = []
lst_100to999 = []
lst_1000to_9999 = []

for i in range(0, 10000):
    if i < 10:
        lst_0to9.append(file_lst[i])
        continue
    elif i < 100:
        lst_10to99.append(file_lst[i])
        continue        
    elif i < 1000:
        lst_100to999.append(file_lst[i])
        continue                
    elif i < 10000:
        lst_1000to_9999.append(file_lst[i])
        continue      

lst_0to9.sort()
lst_10to99.sort()
lst_100to999.sort()
lst_1000to_9999.sort()

lst_total.append(lst_0to9)
lst_total.append(lst_10to99)
lst_total.append(lst_100to999)
lst_total.append(lst_1000to_9999)
# lst_total[0] : 0~9
# lst_total[1] : 10~99
# lst_total[2] : 100~999
# lst_total[3] : 1000~9999
```

```python
char_lst_1 = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M',
              'O', 'P', 'Q', 'R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y', 'Z']
char_lst_2 = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm',
              'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z',]
char_lst_3 = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0']
char_lst_4 = ['_', '+', '`', '^', '?', '|', ';', '-', '~', "/", ":"]
char_lst_5 = ['!', '#', '$', '%', '&', '*', '=', '(', ')', '{', '}', '[', ']']


def make_rand_lst(char_lst_1, char_lst_2, char_lst_3, char_lst_4, char_lst_5):
    rand_lst = []
    rand_lst_1 = random.sample(char_lst_1, 1)
    rand_lst_2 = random.sample(char_lst_2, 1)
    rand_lst_3 = random.sample(char_lst_3, 1)
    rand_lst_4 = random.sample(char_lst_4, 4)
    rand_lst_5 = random.sample(char_lst_5, 2)
    rand_lst.append(" ")
    rand_lst = rand_lst + rand_lst_1 + rand_lst_2 + rand_lst_3 + rand_lst_4 + rand_lst_5
    return rand_lst

def to_greyscale(image):
    return image.convert("L")

def to_greyscale(image):
    return image.convert("L")

def pixel_to_ascii(image):
    
    ASCII_CHARS = [" ", ".", "-", "/", ";", "c", "1", "2", "0", "@"]
#     ASCII_CHARS = make_rand_lst(char_lst_1, char_lst_2, char_lst_3, char_lst_4, char_lst_5)
    
    pixels = image.getdata()
    ascii_str = "";
    for pixel in pixels:
        ascii_str += ASCII_CHARS[pixel//26];
    return ascii_str

def resize(image, new_width = 100):
    width, height = image.size
    new_height = new_width * height / width * 0.5
    return image.resize((int(new_width), int(new_height)))

def convert_and_save(path, save_path, file_name):
    try:
        image = Image.open(path)
    except:
        print(path, "Unable to find image ")

    image = resize(image)

    greyscale_image = to_greyscale(image)

    ascii_str = pixel_to_ascii(greyscale_image)
    img_width = greyscale_image.width
    ascii_str_len = len(ascii_str)
    ascii_img=""

    for i in range(0, ascii_str_len, img_width):
        ascii_img += ascii_str[i:i+img_width] + "\n"

    with open(save_path+"/"+file_name, "w") as f:
        f.write(ascii_img);
```

```python
path = "/Users/jinsungpark/Desktop/crypto_total"
save_path = "/Users/jinsungpark/Desktop/crypto_ascii"
print("percentage   0                      25                      50                        75                      100")
print("converting   ", end='')
for i in range(0, 10000):
#     print("{}/10000   ".format(i), end='')
    if i%100 == 0:
        print("|", end='')
    if i < 10:
        convert_and_save(path+"/"+lst_total[0][i], save_path+"/"+"file_0-999", lst_total[0][i][:13]+".txt")
        continue
    elif i < 100:
        convert_and_save(path+"/"+lst_total[1][i-10], save_path+"/"+"file_0-999", lst_total[1][i-10][:14]+".txt")
        continue
    elif i < 1000:
        convert_and_save(path+"/"+lst_total[2][i-100], save_path+"/"+"file_0-999", lst_total[2][i-100][:15]+".txt")
        continue
    elif i < 2000:
        convert_and_save(path+"/"+lst_total[3][i-1000], save_path+"/"+"file_1000-1999", lst_total[3][i-1000][:16]+".txt")        
        continue
    elif i < 3000:
        convert_and_save(path+"/"+lst_total[3][i-1000], save_path+"/"+"file_2000-2999", lst_total[3][i-1000][:16]+".txt")        
        continue
    elif i < 4000:
        convert_and_save(path+"/"+lst_total[3][i-1000], save_path+"/"+"file_3000-3999", lst_total[3][i-1000][:16]+".txt")                
        continue
    elif i < 5000:
        convert_and_save(path+"/"+lst_total[3][i-1000], save_path+"/"+"file_4000-4999", lst_total[3][i-1000][:16]+".txt")                
        continue
    elif i < 6000:
        convert_and_save(path+"/"+lst_total[3][i-1000], save_path+"/"+"file_5000-5999", lst_total[3][i-1000][:16]+".txt")                
        continue
    elif i < 7000:
        convert_and_save(path+"/"+lst_total[3][i-1000], save_path+"/"+"file_6000-6999", lst_total[3][i-1000][:16]+".txt")                
        continue
    elif i < 8000:
        convert_and_save(path+"/"+lst_total[3][i-1000], save_path+"/"+"file_7000-7999", lst_total[3][i-1000][:16]+".txt")                
        continue
    elif i < 9000:
        convert_and_save(path+"/"+lst_total[3][i-1000], save_path+"/"+"file_8000-8999", lst_total[3][i-1000][:16]+".txt")             
        continue
    elif i < 10000:
        convert_and_save(path+"/"+lst_total[3][i-1000], save_path+"/"+"file_9000-9999", lst_total[3][i-1000][:16]+".txt")                
        continue
```
